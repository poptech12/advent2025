```python
# get length of integers in range
# get range
# look within that number to find any repeating numbers 
  # find mind and max length of digits
  # i.e., if the numbers are just 1-2 digits, then you just look for two repeating numbers
  # if the number is 2-3 digits long you look for 2 or 3 repeating numbers, etc
  # if the max length and min length are equal then you can just split that number in half - 
  # e.g., look for one pattern in first half that matches a sequence in second half.
  # you onlt need to look for lengths divisible by 2 (2, 4, 6, 8, 10, 12, etc.)
import re

def find_repeating_number_sequences(text):
    """
    Finds all repeating sequences of numbers within a given string.

    Args:
        text (str): The input string to search.

    Returns:
        list: A list of tuples, where each tuple contains the repeating sequence
              and the number of times it repeats.
    """
    # Regex explanation:
    # (\d+): Captures one or more digits (the repeating sequence).
    # \1+: Matches one or more occurrences of the captured group.
    digit_length = len(text)
    pattern = r"(\d+)\1+"
    matches = re.finditer(pattern, text)
    results = []
    for match in matches:
        sequence = match.group(1)
        # Calculate the number of repetitions
        repetitions = len(match.group(0)) // len(sequence)
        results.append((sequence, repetitions))
    return results

with open("inputDay2example.txt", "r") as file:
    indiv_numbers = file.readline().split(',')
    # print (indiv_numbers)
    for indiv_number in indiv_numbers :
        start_range = re.match(r"^(\w+)-", indiv_number)
        end_range =  re.search(r"-(\w+)", indiv_number)
        start_range_clean = start_range.group(0).replace("-","")
        end_range_clean = end_range.group(0).replace("-","")
        # print ('line: '+line)
        start_range_length = len(start_range_clean)
        end_range_length = len(end_range_clean)
        print ("start_range_length: "+str(start_range_length))
        print ("end_range_length: "+str(end_range_length))
        range_count = int(end_range_clean)-int(start_range_clean)
        if (start_range_length == end_range_length) :
            print ("Range lengths match!")
            half_length = start_range_length / 2
            print ("half_length: "+str(half_length))
            # print(start_range_clean)
            # print(find_repeating_number_sequences(start_range_clean))
            for i in range(int(start_range_clean),int(end_range_clean)+1) :
                print (i)
                num_as_string = str(i)
                if find_repeating_number_sequences(num_as_string):
                  print(find_repeating_number_sequences(num_as_string))

        else :
            print ("Range lengths don't match.")
        print("range_count: "+str(range_count)+"\n")
```
